```python
import heapq

# Grafo representado como diccionario
graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 2, 'E': 5},
    'C': {'F': 3},
    'D': {'G': 1},
    'E': {'G': 2},
    'F': {'G': 5},
    'G': {}
}

# Heurística estimada al objetivo
heuristic = {
    'A': 7,
    'B': 6,
    'C': 5,
    'D': 2,
    'E': 1,
    'F': 3,
    'G': 0
}

def a_star(start, goal):
    queue = [(0, start)]
    costs = {start: 0}
    path = {}

    while queue:
        _, current = heapq.heappop(queue)

        if current == goal:
            break

        for neighbor, weight in graph[current].items():
            new_cost = costs[current] + weight

            if neighbor not in costs or new_cost < costs[neighbor]:
                costs[neighbor] = new_cost
                priority = new_cost + heuristic[neighbor]
                heapq.heappush(queue, (priority, neighbor))
                path[neighbor] = current

    # reconstruir camino
    current = goal
    route = [current]

    while current != start:
        current = path[current]
        route.append(current)

    route.reverse()
    return route

print("Ruta encontrada:", a_star('A','G'))
```

```python
import random

N = 8

def conflicts(board):
    count = 0
    for i in range(N):
        for j in range(i+1, N):
            if board[i] == board[j] or abs(board[i]-board[j]) == j-i:
                count += 1
    return count

def random_board():
    return [random.randint(0, N-1) for _ in range(N)]

def hill_climbing():
    board = random_board()

    while True:
        current_conflicts = conflicts(board)
        best_board = board
        best_conflicts = current_conflicts

        for col in range(N):
            for row in range(N):
                new_board = board.copy()
                new_board[col] = row
                new_conflicts = conflicts(new_board)

                if new_conflicts < best_conflicts:
                    best_board = new_board
                    best_conflicts = new_conflicts

        if best_conflicts >= current_conflicts:
            return board

        board = best_board

solution = hill_climbing()
print("Solución:", solution)
print("Conflictos:", conflicts(solution))
```
